```python
import pandas as pd
import neurokit2 as nk
```

```python
amel = pd.read_csv('amel_tkr.csv',sep = ';')
andika = pd.read_csv('andika_tkr.csv',sep = ';')
dhea = pd.read_csv('dhea_tkr.csv',sep = ';')
elya = pd.read_csv('elya_tkr.csv',sep = ';')
fauzi = pd.read_csv('fauzi_tkr.csv',sep = ';')
iam = pd.read_csv('iam_tkr.csv',sep = ';')
kanwa = pd.read_csv('kanwa_tkr.csv',sep = ';')
nanda = pd.read_csv('nanda_tkr.csv',sep = ';')
nandaeka = pd.read_csv('nandakkn_tkr.csv',sep = ';')
nopen = pd.read_csv('nopen_tkr.csv',sep = ';')
nopen2 = pd.read_csv('nopen_tkr2.csv',sep = ';')
pradja = pd.read_csv('pradja_tkr.csv',sep = ';')
raihan = pd.read_csv('raihan_tkr.csv',sep = ';')
rico = pd.read_csv('rico_tkr.csv',sep = ';')
silvia = pd.read_csv('silvia_tkr.csv',sep = ';')
visky = pd.read_csv('visky_tkr.csv',sep = ';')
```

```python
amel.columns = ["ir","waktu"]
andika.columns = ["ir","waktu"]
dhea.columns = ["ir","waktu"]
elya.columns = ["ir","waktu"]
fauzi.columns = ["ir","waktu"]
iam.columns = ["ir","waktu"]
kanwa.columns = ["ir","waktu"]
nanda.columns = ["ir","waktu"]
nandaeka.columns = ["ir","waktu"]
nopen.columns = ["ir","waktu"]
nopen2.columns = ["ir","waktu"]
pradja.columns = ["ir","waktu"]
raihan.columns = ["ir","waktu"]
rico.columns = ["ir","waktu"]
silvia.columns = ["ir","waktu"]
visky.columns = ["ir","waktu"]
```

```python
amel = amel[(amel.waktu <= 60)]
andika = andika[(andika.waktu <= 60)]
dhea = dhea[(dhea.waktu <= 60)]
elya = elya[(elya.waktu <= 60)]
fauzi = fauzi[(fauzi.waktu <= 60)]
iam = iam[(iam.waktu <= 60)]
kanwa = kanwa[(kanwa.waktu <= 60)]
nanda = nanda[(nanda.waktu <= 60)]
nandaeka = nandaeka[(nandaeka.waktu <= 60)]
nopen = nopen[(nopen.waktu <= 60)]
nopen2 = nopen2[(nopen2.waktu <= 60)]
pradja = pradja[(pradja.waktu <= 60)]
raihan = raihan[(raihan.waktu <= 60)]
rico = rico[(rico.waktu <= 60)]
silvia = silvia[(silvia.waktu <= 60)]
visky = visky[(visky.waktu <= 60)]
```

```python
ppg_amel = nk.ppg_clean(amel[['ir']], sampling_rate=200)
peaks_amel = nk.ppg_findpeaks(ppg_amel,sampling_rate=35,show=False)

ppg_andika = nk.ppg_clean(andika[['ir']], sampling_rate=200)
peaks_andika = nk.ppg_findpeaks(ppg_andika,sampling_rate=35,show=False)

ppg_dhea = nk.ppg_clean(dhea[['ir']], sampling_rate=200)
peaks_dhea = nk.ppg_findpeaks(ppg_dhea,sampling_rate=35,show=False)

ppg_elya = nk.ppg_clean(elya[['ir']], sampling_rate=200)
peaks_elya = nk.ppg_findpeaks(ppg_elya,sampling_rate=35,show=False)

ppg_fauzi = nk.ppg_clean(fauzi[['ir']], sampling_rate=200)
peaks_fauzi = nk.ppg_findpeaks(ppg_fauzi,sampling_rate=35,show=False)

ppg_iam = nk.ppg_clean(iam[['ir']], sampling_rate=200)
peaks_iam = nk.ppg_findpeaks(ppg_iam,sampling_rate=35,show=False)

ppg_kanwa = nk.ppg_clean(kanwa[['ir']], sampling_rate=200)
peaks_kanwa = nk.ppg_findpeaks(ppg_kanwa,sampling_rate=35,show=False)

ppg_nanda = nk.ppg_clean(nanda[['ir']], sampling_rate=200)
peaks_nanda = nk.ppg_findpeaks(ppg_nanda,sampling_rate=35,show=False)

ppg_nandaeka = nk.ppg_clean(nandaeka[['ir']], sampling_rate=200)
peaks_nandaeka = nk.ppg_findpeaks(ppg_nandaeka,sampling_rate=35,show=False)

ppg_nopen = nk.ppg_clean(nopen[['ir']], sampling_rate=200)
peaks_nopen = nk.ppg_findpeaks(ppg_nopen,sampling_rate=35,show=False)

ppg_nopen2 = nk.ppg_clean(nopen2[['ir']], sampling_rate=200)
peaks_nopen2 = nk.ppg_findpeaks(ppg_nopen2,sampling_rate=35,show=False)

ppg_pradja = nk.ppg_clean(pradja[['ir']], sampling_rate=200)
peaks_pradja = nk.ppg_findpeaks(ppg_pradja,sampling_rate=35,show=False)

ppg_raihan = nk.ppg_clean(raihan[['ir']], sampling_rate=200)
peaks_raihan = nk.ppg_findpeaks(ppg_raihan,sampling_rate=35,show=False)

ppg_rico = nk.ppg_clean(rico[['ir']], sampling_rate=200)
peaks_rico = nk.ppg_findpeaks(ppg_rico,sampling_rate=35,show=False)

ppg_silvia = nk.ppg_clean(silvia[['ir']], sampling_rate=200)
peaks_silvia = nk.ppg_findpeaks(ppg_silvia,sampling_rate=35,show=False)

ppg_visky = nk.ppg_clean(visky[['ir']], sampling_rate=200)
peaks_visky = nk.ppg_findpeaks(ppg_visky,sampling_rate=35,show=False)
```
